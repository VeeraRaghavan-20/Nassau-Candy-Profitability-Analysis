## Profit contribution percentage

```python
profit_contribution = (Candy.groupby('Product Name')['Gross Profit'].sum() / Candy['Gross Profit'].sum()) * 100
print(profit_contribution.sort_values(ascending=False))
```

## Cost Structure Diagnostics


```python
import matplotlib.pyplot as plt
plt.scatter(Candy['Cost'], Candy['Sales'])
plt.show()
```

```python
import matplotlib.pyplot as plt

colors = {'Chocolate': 'brown', 'Sugar': 'blue', 'Other': 'green'}

for division, group in Candy.groupby('Division'):
    plt.scatter(group['Cost'], group['Sales'], 
                label=division, 
                color=colors[division],
                alpha=0.6)

plt.xlabel('Cost')
plt.ylabel('Sales')
plt.title('Cost vs Sales by Division')
plt.legend()
plt.show()
```

## Flag risk products

```python
import numpy as np

conditions = [
    Candy['Gross Margin'] < 40,
    (Candy['Gross Margin'] >= 40) & (Candy['Gross Margin'] < 60),
    Candy['Gross Margin'] >= 60
]

labels = ['Discontinue Review', 'Reprice', 'Healthy']

Candy['Risk Flag'] = np.select(conditions, labels, default='Unknown')

print(Candy['Risk Flag'].value_counts())
```
